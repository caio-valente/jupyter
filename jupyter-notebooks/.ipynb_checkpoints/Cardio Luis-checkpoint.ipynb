```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio #graph output

import pandas as pd
pd.options.plotting.backend = "plotly"
import numpy as np

from ipywidgets import interact_manual
```

```python
df = pd.read_excel(r'C:\Users\Caio\Downloads\Instrumento de Coleta de dados - IC (respostas).xlsx')
```

```python
del df['dfg']
```

```python
del df['Unnamed: 34']
```

```python
df = df.dropna(axis = 'rows')
```

```python
df.columns
```

```python
def plot_scatter(X,Y,C): 
    fig = df.plot(kind = 'scatter', y = Y, x = X, color = C)
    fig.show()
interact_manual(plot_scatter,X=df.columns,Y=df.columns,C=df.columns)
```
