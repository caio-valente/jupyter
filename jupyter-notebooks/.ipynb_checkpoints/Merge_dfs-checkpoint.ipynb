```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go


def triboscopy(dff):
    fig = px.scatter_3d(dff, x='X',y='T',z='COF', color='COF')
    fig.update_traces(marker_size = 5)
    fig.update_layout(margin=dict(l=20, r=20, t=40, b=0),
                      scene_camera_eye=dict(x=1.4, y=1.4, z=1.4),
                      scene_xaxis_title_text = 'Posição (mm)',
                      scene_yaxis_title_text = 'Tempo (s)',
                      scene_zaxis_title_text = 'COF (-)',
                      )
    return fig
```

```python
file = r'C:\Users\Caio\OneDrive\Documentos\Python\Results\CSV\GPP11\Umidade_FC_001.csv'
df = pd.read_csv(file,header=17, skiprows = [21], dtype = np.float64).filter(items=['T','Fz','COF'])
```

```python
file = r'C:\Users\Caio\OneDrive\Documentos\Python\Results\CSV\GPP11\Umidade_FC_002.csv'
df2 = pd.read_csv(file,header=17, skiprows = [21], dtype = np.float64).filter(items=['T','Fz','COF'])
```

```python
df = df.head(len(df)- 15000)
```

```python
zero = df.tail(n = 1000).nsmallest(1,'COF').iat[0,0]
df = df.assign(V = - 4.66 * np.pi * 2 * np.sin((2.*np.pi*2 *(df['T'] - zero))),
                 X = 4.66/2 * np.cos(2.*np.pi*2*(df['T']-zero)))
df['Velocidade instantânea'] = df['V']>= 2 * 2 * 4.66
```

```python
df2 = df2.tail(len(df2)-1000)
zero = df2.tail(n = 1000).nsmallest(1,'COF').iat[0,0]
df2 = df2.assign(V = - 4.66 * np.pi * 2 * np.sin((2.*np.pi*2 *(df2['T'] - zero))),
                 X = -4.66/2 * np.cos(2.*np.pi*2*(df2['T']-zero)))
df2['Velocidade instantânea'] = df['V']>= 2 * 2 * 4.66
df2['T'] = df2['T']+df.tail(1).iat[0,0]-1
```

```python
df = df.append(df2, ignore_index = True)
```

```python
triboscopy(df.sample(frac = 0.01))
```

```python
df.to_csv('Umidade_FC_001.csv')
```
