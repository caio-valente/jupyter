```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_palette("colorblind")
```

```python
df = pd.read_csv(r'C:\Users\Caio\OneDrive\Documentos\Python\Results\CSV\Compilado\reciprocating_anova.csv')
```

```python
df.drop(['Unnamed: 0','indice','Intercept','COF Desv. Pad.'], axis = 'columns', inplace = True)
```

```python
data = df[['Material',
          'Frequência',
          'Força Normal',
          'EG',
#           'Frequência²',
#           'Frequência * EG',
#           'Força * EG',
#           'Força * Frequência',
#           'Força * Frequência * EG',
          'COF Médio',
          'Largura Média'
          ]]
```

```python
sns.pairplot(data, hue = 'Material',corner = True)
```

```python
for material in data['Material'].unique():
    if data[data['Material']==material]['Largura Média'].dropna().size>0:
        plt = sns.jointplot(x="Largura Média", y="COF Médio", kind="reg",data=data[data['Material']==material])
        plt.fig.suptitle(material, y = 1.05)
        plt
plt = sns.jointplot(x="Largura Média", y="COF Médio", kind="reg",data=data)
plt.fig.suptitle('Combinado', y = 1.05)
plt
```
