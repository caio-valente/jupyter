```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots

import pandas as pd
pd.options.plotting.backend = "plotly"
import numpy as np
from ipywidgets import interact, widgets
```

```python
filenames = ['PMoida.txt', 'PNMoida.txt']
frames = dict()
for filename in filenames:
        filepath = 'src/'+filename
        try:
            frames[filepath] = pd.read_csv(filepath,
                                                delim_whitespace=True,
                                                skiprows=32,
                                                names=['2Theta','I'])
            frames[filepath]['I_norm'] = frames[filepath]['I']/frames[filepath]['I'].max()
        except FileNotFoundError:
            pass
```

```python
interact(lambda key: px.line(frames[key],x = '2Theta', y='I_norm'),
         key = list(frames.keys())
        )
```

```python
import pywt
from statsmodels.robust import mad
cA, cD = pywt.dwt(frames['src/PMoida.txt']['I_norm'], 'db2')
```

```python
px.scatter(x=cA, y=cD)
```

```python
df = pd.DataFrame({'cA':cA,'cD':cD})
```

```python
i = pywt.idwt(df[df['cA']<=1]['cA'], df[df['cA']<=1]['cD'], 'db2')
```

```python
px.line(y = i)
```

```python
x = frames['src/PMoida.txt']
coef = pywt.wavedec(x['I'], 'db2', mode='per')
```

```python
level = 4
sigma = mad( coef[-level] )
uthresh = sigma * np.sqrt( 2*np.log( len( x['I'] ) ) )
coef[1:] = ( pywt.threshold( i, value=uthresh, mode="soft" ) for i in coef[1:] )
```

```python
y = pywt.waverec( coef, 'db2', mode="per" )
```

```python
px.line(y=y/max(y), x=x['2Theta'])
```
